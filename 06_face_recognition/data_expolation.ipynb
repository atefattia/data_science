# Explore dataset


## Names

```python
import pandas as pd
import numpy as np
```

```python
!head /storage/lfw/lfw_names.txt
```

```python
names = pd.read_csv('./storage/lfw/lfw_names.txt', sep='\t', header=None, names=['name', 'num_images'])
names['start_index'] = names['num_images'].cumsum().shift(1, fill_value=0)
names['name'] = names['name'].apply(lambda x: x.replace('_', ' '))
names.set_index('name', inplace=True)
names
```

## Attributes

```python
!head ./storage/lfw/lfw_attributes.txt -n 3
```

```python
# read first line
with open('./storage/lfw/lfw_attributes.txt') as file:
    next(file)
    headRow = next(file)

columns = [x.strip() for x in headRow.split('\t')]

attributes = pd.read_csv('./storage/lfw/lfw_attributes.txt', sep='\t', names=columns[1:], skiprows=2)
attributes
```

some images are missing attributes because the sizes don't match

```python
(names.iloc[-1,1] + names.iloc[-1,0]) - attributes.shape[0]
```

```python
tmp = names.copy()
tmp['imagenum'] = tmp['num_images'].apply(lambda x: list(range(x)))
tmp = tmp.explode('imagenum')
tmp['imagenum'] = tmp['imagenum'] + 1
tmp.reset_index(inplace=True)
tmp.rename(columns={'name': 'person'}, inplace=True)
merge = tmp.merge(attributes, how='outer', on=['person', 'imagenum'], sort=True, indicator=True)
merge[merge['_merge'] != 'both']
merge
```

fix attributes and fill missing attributes with NaN

```python
attributes_new = merge
```

```python
names.to_csv('/storage/lfw/lfw_names.csv')
attributes_new.to_csv('/storage/lfw/lfw_attributes.csv')
```
